```python
import pandas as pd
from pathlib import Path

PROJECT_DIR = Path.cwd()
DATA_DIR = PROJECT_DIR / "project_data"
```

```python
possesion_data = pd.read_csv(DATA_DIR / 'player_possession.csv')

# Select only midfielders from the 22/23 season
mf_possesion_data = possesion_data[(possesion_data['Season_End_Year'] == 2023) & (possesion_data['Pos'].str.contains('MF', na=False))].drop(columns='Url')

# Check for missing data
# mf_possesion_data.isna().sum()

# Remove columns that contains only NaN values
mf_possesion_data_no_nan_col = mf_possesion_data[[col for col in mf_possesion_data.columns if mf_possesion_data[col].isna().sum() != 1202]]
mf_possesion_data_no_nan_col

# Check
# mf_possesion_data_no_nan_col.isna().sum()

# Drop columns that is calculated from other columns (e.g., percentages)
mf_possesion_data_no_redundant = mf_possesion_data_no_nan_col[[col for col in mf_possesion_data_no_nan_col.columns if '_percent_' not in col]]
mf_possesion_data_no_redundant
```

```python
mf_possesion_data_no_redundant.isna().sum()
```

```python
passing_data = pd.read_csv(DATA_DIR / 'player_passing.csv')

mf_passing_data = passing_data[(passing_data['Season_End_Year'] == 2023) & (passing_data['Pos'].str.contains('MF', na=False))].drop(columns='Url')

# Check for missing data
# mf_passing_data.isna().sum()

# Remove columns that contains only NaN values
mf_passing_data_no_nan_col = mf_passing_data[[col for col in mf_passing_data.columns if mf_passing_data[col].isna().sum() != 1202]]
mf_passing_data_no_nan_col

# Check
# mf_passing_data_no_nan_col.isna().sum()

# Drop columns that is calculated from other columns (e.g., percentages)
mf_passing_data_no_redundant = mf_passing_data_no_nan_col[[col for col in mf_passing_data_no_nan_col.columns if '_percent_' not in col]]
mf_passing_data_no_redundant
```

```python
mf_passing_data_no_redundant.isna().sum()
```

```python
shooting_data = pd.read_csv(DATA_DIR / 'player_shooting.csv', low_memory=False)

mf_shooting_data = shooting_data[(shooting_data['Season_End_Year'] == 2023) & (shooting_data['Pos'].str.contains('MF', na=False))].drop(columns='Url')

# Drop columns that is calculated from other columns (e.g., percentages)
mf_shooting_data_no_redundant = mf_shooting_data[[col for col in mf_shooting_data.columns if '_percent_' not in col]]

# Check that there are still columns with many missing data
# mf_shooting_data_no_redundant.isna().sum()

# Drop those columns
mf_shooting_data_no_redundant = mf_shooting_data_no_redundant.drop(columns=['G_per_Sh_Standard', 'G_per_SoT_Standard', 'Dist_Standard', 'npxG_per_Sh_Expected'])
mf_shooting_data_no_redundant
```

```python
mf_shooting_data_no_redundant.isna().sum()
```

```python
defense_data = pd.read_csv(DATA_DIR / 'player_defense.csv')

mf_defense_data = defense_data[(defense_data['Season_End_Year'] == 2023) & (defense_data['Pos'].str.contains('MF', na=False))].drop(columns='Url')

# Check for missing data
# mf_defense_data.isna().sum()

# Remove columns that contains only NaN values
mf_defense_data_no_nan_col = mf_defense_data[[col for col in mf_defense_data.columns if mf_defense_data[col].isna().sum() != 1202]]
mf_defense_data_no_nan_col

# Drop columns that is calculated from other columns (e.g., percentages)
mf_defense_data_no_redundant = mf_defense_data_no_nan_col[[col for col in mf_defense_data_no_nan_col.columns if '_percent_' not in col]]
mf_defense_data_no_redundant
```

```python
mf_defense_data_no_redundant.isna().sum()
```

```python
gca_data = pd.read_csv(DATA_DIR / 'player_gca.csv')

mf_gca_data = gca_data[(gca_data['Season_End_Year'] == 2023) & (gca_data['Pos'].str.contains('MF', na=False))].drop(columns='Url')
mf_gca_data
```

```python
mf_gca_data.isna().sum()
```

```python
standard_data = pd.read_csv(DATA_DIR / 'player_standard.csv', low_memory=False)

drop_cols = ['Pos', 'Age', 'Url', 'xG_Expected', 'npxG_Expected', 'Ast']

mf_standard_data = standard_data[(standard_data['Season_End_Year'] == 2023) & (standard_data['Pos'].str.contains('MF', na=False))].drop(columns=drop_cols)
mf_standard_data
```

```python
mf_standard_data.isna().sum()
```

```python
# Merge all dataframes on common identifier columns
merged_data = mf_possesion_data_no_redundant.merge(
    mf_passing_data_no_redundant, 
    on=['Player', 'Season_End_Year', 'Squad', 'Nation', 'Pos', 'Age', 'Born', 'Comp', 'Mins_Per_90'], 
    how='outer'
).merge(
    mf_shooting_data_no_redundant, 
    on=['Player', 'Season_End_Year', 'Squad', 'Nation', 'Pos', 'Age', 'Born', 'Comp', 'Mins_Per_90'], 
    how='outer'
).merge(
    mf_defense_data_no_redundant, 
    on=['Player', 'Season_End_Year', 'Squad', 'Nation', 'Pos', 'Age', 'Born', 'Comp', 'Mins_Per_90'], 
    how='outer'
).merge(
    mf_gca_data, 
    on=['Player', 'Season_End_Year', 'Squad', 'Nation', 'Pos', 'Age', 'Born', 'Comp', 'Mins_Per_90'], 
    how='outer'
).merge(
    mf_standard_data,
    on=['Player', 'Season_End_Year','Squad', 'Nation', 'Comp', 'Born'],
    how='inner'
    )

merged_data.drop_duplicates(inplace=True)
```

```python
merged_data.columns.tolist()
```

```python
merged_data = merged_data.dropna()
merged_data
```

```python
merged_data.isna().sum().sum()
```

```python
merged_data.to_csv(PROJECT_DIR / 'mf_data_europe_2023.csv', index=False)
```
